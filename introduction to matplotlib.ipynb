# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3]
y = [11,22,33]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple plot", xlabel="x-axis", ylabel="y-axis")

# 5. save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

```python
# Make a plot from dictionary
nut_butter_price = {"Almont butter":10, "Peanuts butter": 8, "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_price.keys(), nut_butter_price.values())
ax.set(title = "Dan's Nut Butter Store", ylabel = "Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_price.keys()), list(nut_butter_price.values()))
```

```python
# Make some data for histograms
x = np.random.randn(10000)

fig, ax = plt.subplots()
ax.hist(x);
```

# Two options for subplots

```python
# option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5));

# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_price.keys(), nut_butter_price.values());
ax4.hist(np.random.randn(1000));
```

```python
# option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5));

# plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_price.keys(), nut_butter_price.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
# Let's try on another dataset
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# Create a histogram
heart_disease["age"].plot.hist(bins=50);
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10,30));
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.plot(kind="scatter", x="age", y='chol', c='target', figsize=(10,6));
```

```python
## OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x='age', y='chol', c='target', ax=ax);
ax.set_xlim([45,100]);
```

```python
## OO methode from scratch
fig, ax = plt.subplots(figsize=(10,6))

## Plot the data
scatter = ax.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"]
);

# Customize the plot
ax.set(title="Heart Diseases and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol");

# Add legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
over_50.head()
```

```python
## Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)

## Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50['chol'], c=over_50["target"]);

## Customiza ax0
ax0.set(title="Heart Diseases and Cholesterol Levels", ylabel="Cholesterol");

## Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");

## Add data to ax1
scatter = ax1.scatter(x=over_50["age"], y=over_50['thalach'], c=over_50["target"]);

## Customiza ax01
ax1.set(title="Heart Diseases and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate");

## Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

## Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# set the style
plt.style.use('seaborn-v0_8-whitegrid')

## Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)

## Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50['chol'], c=over_50["target"], cmap="winter");

## Customiza ax0
ax0.set(title="Heart Diseases and Cholesterol Levels", ylabel="Cholesterol");

## Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");

## Add data to ax1
scatter = ax1.scatter(x=over_50["age"], y=over_50['thalach'], c=over_50["target"]);

## Customiza ax01
ax1.set(title="Heart Diseases and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate");

## Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

## Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
## Customizing axis limitations 

## Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)

## Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50['chol'], c=over_50["target"], cmap="winter");

## Customiza ax0
ax0.set(title="Heart Diseases and Cholesterol Levels", ylabel="Cholesterol");

## Change x axis limits
ax0.set_xlim([50, 80]);

## Change y axis limits
ax0.set_ylim([100, 600]);

## Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");

## Add data to ax1
scatter = ax1.scatter(x=over_50["age"], y=over_50['thalach'], c=over_50["target"], cmap="winter");

## Customiza ax01
ax1.set(title="Heart Diseases and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate");

## Change x axis limits
ax1.set_xlim([50,80])

## Change y axis limits
ax1.set_ylim([60, 200]);

## Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

## Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
